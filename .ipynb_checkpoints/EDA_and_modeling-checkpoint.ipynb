Given that we had so many null values in the brand column we need to come up with an efficient way to sort through those tweets so that we can classify them as one of the brands or products already provided by the original dataset.  Trying to manually comb through 5000 tweets is far too time consuming so instead we will iterate through the unbranded tweets and search for keywords within the text.

In this case we will use nested for loops to iterate through each row of the dataframe and each word in a tweet. We will have a list of keywords/phrases that point to a particular brand or product. Here the keywords/phrases will just be the brands/products already listed in the dataset. We will change the capitalization of the keywords and phrases in hopes that we can capture as many unbranded tweets as possible. If the keyword or phrase is in the tweet, then we will reassign that tweet from `no_brand` to one of the brands/products already listed.

The order in which we run these loops does matter. For example if a tweet were to reference both an iPhone and and an iPad, executing a loop with the iPhone keywords before the iPad keywords would cause that tweet to be labeled as an iPhone related tweet. Given that the capitalizations are varied and we don't know the order of words in the unbranded tweets, there are far too many different combinations that would need to be tried. A for loop going through each of the combinations would also require many complex `if-else` statements. The main takeaway here is that we want to be efficient and generalizable. 

```python
ipad_app_keywords = ['iPad app', 'ipad app', 'IPad app', 'IPAD app', 'Ipad app']
iphone_app_keywords = ['iPhone app', 'iphone app', 'IPhone app', 'Iphone app', 'IPHONE app']
android_app_keywords = ['android app', 'Android app', 'ANDROID app']

#reclassify unbranded tweets as tweets related to iOS apps
for i in range(0, len(df)):                       #iterate through each row in the dataframe
    if df.loc[i, 'brand_product'] == 'no_brand':  #only select unbranded tweets
        tweet_phrase = df.loc[i, 'tweet']         #store the tweet string in a variable
        for word in ipad_app_keywords:            #iterate through each word in the respective keywords list
            if word in tweet_phrase:          #use if-else statements to check for keywords and reassign if True
                df.loc[i, 'brand_product'] = 'iPad or iPhone App'
            else:                                 #if word doesn't match move to next word
                continue
    else:                                         #skip all tweets that already have a brand or comment
        continue
        

#reclassify unbranded tweets as tweets related to iOS apps
for i in range(0, len(df)):
    if df.loc[i, 'brand_product'] == 'no_brand':
        tweet_phrase = df.loc[i, 'tweet']
        for word in iphone_app_keywords:
            if word in tweet_phrase:
                df.loc[i, 'brand_product'] = 'iPad or iPhone App'
            else:
                continue
    else:
        continue
        
#reclassify unbranded tweets as tweets related to Android apps
for i in range(0, len(df)):
    if df.loc[i, 'brand_product'] == 'no_brand':
        tweet_phrase = df.loc[i, 'tweet']
        for word in android_app_keywords:
            if word in tweet_phrase:
                df.loc[i, 'brand_product'] = 'Android App'
            else:
                continue
    else:
        continue
```

We will import the english stopwords from NLTK as well as the WordNetLemmatizer to transform words into their root form. We will use the stopwords and lemmatizer in a function to create the word tokens,

```python
#nltk.download('wordnet')
#nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('english')
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

Below I will create two functions:
1. The first function will simply clean the tweet as we did above so that it appears as it would when using Twitter. This will be useful for EDA and possible feature engineering as we might want to count the number of mentions or hashtags and see if those quantities differ between sentiment.

2. The second function will perform all the same as the first but take it a step further by removing all punctuation and then lemmatizing the words. The tokens will be stored as a list in a separate column.

```python
def clean_tweets(text):
    """This function takes in a string (in this case a tweet) and cleans the string such that any and all
    non-ASCII charcaters and HTML characters references are removed.
    
    The string will still have all forms of punctuation.
    
    """
    text = re.sub(' +', ' ', text) #reduce instances of multiple blank spaces to a single space
    encoded_text = text.encode('ascii', 'ignore') #encode string using 'ascii' encoding
    decode_text = encoded_text.decode() #decode string (convert from bytes to string object)
    clean_text = BeautifulSoup(decode_text, 'lxml').text #remove all html encodings such as &amp;, &quot;, etc.
    
    return clean_text
    
```

```python
def clean_tweets_modeling(text):
    """Performs a similar operation as the clean_tweets function except this function will create a list of 
    tokens that can be used for NLP modeling.
    
    For example, a tweet will be transformed such that all non-ASCII characters and HTML encodings are removed.
    Then the function will remove all punctuation from the tweet, split each word in the tweet by white space,
    and finally lemmatize each word in the tweet and group the lemmatized tokens in a list."""
    
    text = re.sub(' +', ' ', text) #reduce instances of multiple spaces to a single space
    encoded_text = text.encode('ascii', 'ignore') #encode text as ascii characters only
    decode_text = encoded_text.decode() #convert bytes back to a string
    clean_text = BeautifulSoup(decode_text, 'lxml').text #remove html tages using BeautifulSoup
    
    no_punc_text = "".join([word for word in clean_text if word not in string.punctuation]) #remove punctuation
    tokens = re.split('\W+', no_punc_text) #split words by in no_punc_text variable by whitespace
    
    #Lemmatize each token and store tokens in a list
    final_tokens = [wnl.lemmatize(word) for word in tokens if word not in stopwords]
    
    return final_tokens    
```

```python
df_branded['tweet'] = df_branded['tweet'].apply(lambda x: clean_tweets(x))
df_branded['tweet_tokens'] = df_branded['tweet'].apply(lambda x: clean_tweets_modeling(x.lower()))
```

## Exploratory Data Analysis 

Let's first take a look at the most common words in each type of tweet (positive, negative, and neutral). Ideally we would have unique words in each type of tweet as that would make it easier for the model to detect sentiment. If we find that there are many common words in each type of tweet we may have to consider labeling those as stopwords as they will detract from the model being able to determine sentiment.

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
```

```python
positive_text = []
negative_text = []
neutral_text = []

#create dataframes for each type of tweet
df_positive = df_branded.loc[df_branded.sentiment == 1]
df_negative = df_branded.loc[df_branded.sentiment == 0]
df_neutral = df_branded.loc[df_branded.sentiment == 2]

for list_element in df_positive['tweet_tokens']: #loop through the tweet_tokens columns
    for word in list_element: 
        positive_text.append(word) #add each word in the tweet_tokens list to a postive words list
        
for list_element in df_negative['tweet_tokens']:
    for word in list_element:
        negative_text.append(word)
        
for list_element in df_neutral['tweet_tokens']:
    for word in list_element:
        neutral_text.append(word)
        
positive_str = ' '.join(positive_text) #create one string of of all positive words for generating a WordCloud
negative_str = ' '.join(negative_text)
neutral_str = ' '.join(neutral_text)
```

```python
from collections import Counter
#create a dictionary with word counts so we can quantify the most common words
pos_dict = Counter(positive_text)
neg_dict = Counter(negative_text)
neu_dict = Counter(neutral_text)
```

```python
#sort the word counts in descending order
pos_dict_sorted = dict(sorted(pos_dict.items(), key=lambda item: item[1], reverse=True))
neg_dict_sorted = dict(sorted(neg_dict.items(), key=lambda item: item[1], reverse=True))
neu_dict_sorted = dict(sorted(neu_dict.items(), key=lambda item: item[1], reverse=True))

#create dataframes of words counts for each type of tweet
pos_text_df = pd.DataFrame.from_dict(pos_dict_sorted, orient='index', columns=['Count'])
pos_text_df.reset_index(inplace=True)
pos_text_df.rename(columns={'index':'Word'}, inplace=True)

neg_text_df = pd.DataFrame.from_dict(neg_dict_sorted, orient='index', columns=['Count'])
neg_text_df.reset_index(inplace=True)
neg_text_df.rename(columns={'index':'Word'}, inplace=True)

neu_text_df = pd.DataFrame.from_dict(neu_dict_sorted, orient='index', columns=['Count'])
neu_text_df.reset_index(inplace=True)
neu_text_df.rename(columns={'index':'Word'}, inplace=True)
```

### Feature Engineering

It is important to note that an NLP model may require more than just tokens in order to make an accurate prediction. Below I will write functions to count the following:
- the number of hashtags
- the number of mentions
- the number of external weblinks
- length of a tweet (total words not characters)
- total punctuation in a tweet (excluding '#' and '@')

```python
def count_hashtags(text):
    """This function will count the number of hashtags in a tweet. It will search each word for the
    # character and return the total number found within a tweet. Only a cleaned tweet should be passed through
    this function."""
    counter = 0
    for word in text:
        if '#' in word:
            counter += 1
        else:
            continue
    return  counter


def count_mentions(text):
    """This function will count the number of mentions in a tweet. It will search each word for the
    @ character and return the total number found within a tweet. Only a cleaned tweet should be passed
    through this function."""
    counter = 0
    for word in text:
        if '@' in word:
            counter += 1
        else:
            continue
    return counter

def count_weblinks(text):
    """This function will count the number of weblinks in a tweet. It will search each word for 
    'http:' or the string '{link}' and return the total number found within a tweet. Only a cleaned tweet
    should be passed through this function."""
    list_of_counts = []
    for word in text.split():
        http_count = word.count('http:')
        link_count = word.count('{link}')
        list_of_counts.extend([http_count, link_count])
    return sum(list_of_counts)

def count_length_of_tweet(text):
    """This function will count the number of words in a tweet. Only a cleaned tweet should be passed through
    this function."""
    words_no_punc = "".join([word for word in text if word not in string.punctuation])
    return len(words_no_punc.split())

def count_punctuation(text):
    """This function will count the the total punctuation in a tweet.  It will NOT count the '#' or '@' symbols
    as those symbols are used for specific functions on Twitter."""
    annotated_list = []
    punctuation_list = list(string.punctuation)
    punctuation_list.remove('#')
    punctuation_list.remove('@')
    modified_punc = punctuation_list
    
    for word in text.split():
        if word.count('http:') > 0:
            continue
        elif word.count('{link}') > 0:
            continue
        else:
            annotated_list.append(word)
    
    list_of_characters = [char for char in "".join(annotated_list) if char in modified_punc]
    
    return len(list_of_characters)
    
    
```

We will add these features to the dataframe and look at their distributions to see if there is any difference across sentiment.

```python
df_branded['hashtag_count'] = df_branded['tweet'].apply(count_hashtags)
df_branded['mention_count'] = df_branded['tweet'].apply(count_mentions)
df_branded['weblink_count'] = df_branded['tweet'].apply(count_weblinks)
df_branded['tweet_length'] = df_branded['tweet'].apply(count_length_of_tweet)
df_branded['punct_count'] = df_branded['tweet'].apply(count_punctuation)
```

## Modeling

For the vanilla models that will be trained and tested below, I will be using TF-IDF vectorization. This type of word vectorization will be more beneficial because as we saw in the word count bar plots and the wordclouds, several of the same words show up in a tweet regardless of sentiment. By using TF-IDF we can better understand the frequency of the word with respect to how many tweets contain the word in question.

We will pass in our already cleaned tokens to the vectorizer to get a sparse matrix with TF-IDF values for each unique word. 

```python
tfidf_vec = TfidfVectorizer()
X_tfidf = tfidf_vec.fit_transform((df_branded['tweet_tokens'].apply(lambda x: " ".join(x))))
```

```python
print(X_tfidf.shape)
```

The number of unique words is slightly greater than the number of tweets we have. This means that overfitting is definitely a possibility for modeling.

```python
#create dataframe of TF-IDF matrix
X_tfidf_df = pd.DataFrame(X_tfidf.toarray())
X_tfidf_df.columns = tfidf_vec.get_feature_names()
df_vanilla = pd.concat([df_branded['sentiment'], X_tfidf_df], axis=1)
```

```python
X = df_vanilla.drop(columns=['sentiment'])
y = df_vanilla['sentiment']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=123)
```

```python
#create training and test sets with added features
X_af = pd.concat([df_branded['hashtag_count'], df_branded['mention_count'], df_branded['weblink_count'],
              df_branded['tweet_length'], X], axis=1)
X_af_train, X_af_test, y_af_train, y_af_test = train_test_split(X_af, y, test_size=0.3, stratify=y,
                                                         random_state=123)
```

For modeling, I will take the same approach for each algorithm:

1. Train and test a vanilla model on just the TF-IDF matrix
2. Peform a grid search with the vanilla model to optimize hyperparameters
3. Train and test a model with the engineered features added to the TF-IDF matrix
4. Perform a grid search on the added features model 

I will compare accuracies across all models as well as take a look at some confusion matrices to see where the model did or did not perform well.

### Oversampling

Another technique that can be used to combat class imbalance, is oversampling. With oversampling, we will resample randome examples of positive and negative tweets to add to our training data set. The idea is that if the model can see repeated examples of negative or positive tweets it may be possible to improve the performance of the model.

The caveat to oversampling is that we may cause our model to overfit because it is not seeing any new data but rather repeated examples of existing training data.

I chose not to undersample due to the fact that undersampling results in a loss of information and there are too few negative tweets to make undersampling worth the risk. Even if we undersampled so that the class imbalance was present but not as severe, we would be removing thousand of examples from the data. In general, more data leads to better results so undersampling is not a feasible option for this dataset.

```python
##undersampling and oversampling techniques
from imblearn.over_sampling import RandomOverSampler
```

```python
y_train.value_counts() #check number of samples by sentiment in original training data
```

The oversampling quantities for positive and negative class are arbitrarily chosen. I wanted to still maintain some hierarchy in terms of quantity (i.e. neutral > positive > negative) and with so few negative samples to begin with, I wanted to try and minimize overfitting. With only 428 negative samples in the training dataset, oversampling means that it is possible to have simply repeated each of the 428 samples 5 times over. Oversampling does select samples at random with replacement, so that is not necessarily a guarantee, but without knowing which samples it's unclear until we finish modeling whether or not the oversampler selected informative tweets.

```python
#Values for oversampling were arbitrarily chosen
oversample = RandomOverSampler(sampling_strategy={0:2000, 1:2500}, random_state=123)
X_over_train, y_over_train = oversample.fit_resample(X_train, y_train)
```

```python
mnb_over = MultinomialNB(alpha=2, fit_prior=False)
mnb_over.fit(X_over_train, y_over_train)
mnb_over_cv_mean = np.mean(cross_val_score(mnb_over, X_over_train, y_over_train, cv=5, scoring='accuracy'))
```
